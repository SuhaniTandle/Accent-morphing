```python
import os
import librosa
import numpy as np
import random
from tqdm import tqdm
from torch.utils.data import Dataset, DataLoader
import torch
import torch.nn as nn
from google.colab import drive

# --- CONFIGURATION ---
dataset_path = "/content/drive/MyDrive/recordings"
input_accent = "arabic"
target_accent = "english"
sr = 16000
n_fft = 1024
hop_length = 256
max_frames = 128
n_mels = 128
batch_size = 8
epochs = 10
drive.mount('/content/drive')

# --- HELPER ---
def fix_length(spec, max_len):
    if spec.shape[1] > max_len:
        return spec[:, :max_len]
    elif spec.shape[1] < max_len:
        pad_width = max_len - spec.shape[1]
        return np.pad(spec, ((0, 0), (0, pad_width)), mode='constant')
    else:
        return spec

# --- DATASET ---
class AccentSameSentenceDataset(Dataset):
    def __init__(self, dataset_path, input_accent, target_accent, sr, n_fft, hop_length, max_frames, n_mels=80):
        self.input_files = self._load_files(dataset_path, input_accent)
        self.target_files = self._load_files(dataset_path, target_accent)
        self.sr = sr
        self.n_fft = n_fft
        self.hop_length = hop_length
        self.max_frames = max_frames
        self.n_mels = n_mels

    def _load_files(self, path, accent):
        accent_path = os.path.join(path, accent)
        return [os.path.join(accent_path, f) for f in os.listdir(accent_path) if f.endswith(".mp3")]

    def __len__(self):
        return len(self.input_files)

    def __getitem__(self, idx):
        input_file = self.input_files[idx]
        target_file = random.choice(self.target_files)

        x_audio, _ = librosa.load(input_file, sr=self.sr)
        y_audio, _ = librosa.load(target_file, sr=self.sr)

        min_len = min(len(x_audio), len(y_audio))
        x_audio = x_audio[:min_len]
        y_audio = y_audio[:min_len]

        X_spec = librosa.feature.melspectrogram(y=x_audio, sr=self.sr, n_fft=self.n_fft, hop_length=self.hop_length, n_mels=self.n_mels)
        Y_spec = librosa.feature.melspectrogram(y=y_audio, sr=self.sr, n_fft=self.n_fft, hop_length=self.hop_length, n_mels=self.n_mels)

        X_log = np.log1p(X_spec)
        Y_log = np.log1p(Y_spec)

        X_log = fix_length(X_log, self.max_frames)
        Y_log = fix_length(Y_log, self.max_frames)

        X_tensor = torch.tensor(X_log, dtype=torch.float32).unsqueeze(0)
        Y_tensor = torch.tensor(Y_log, dtype=torch.float32).unsqueeze(0)

        return X_tensor, Y_tensor

# --- MODEL ---
class AccentMorpherCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
        )
        self.bottleneck = nn.Sequential(
            nn.Conv2d(64, 64, kernel_size=1),
            nn.ReLU()
        )
        self.decoder = nn.Sequential(
            nn.Conv2d(64, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.Conv2d(32, 1, kernel_size=3, padding=1)
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.bottleneck(x)
        x = self.decoder(x)
        return x

# --- TRAINING ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
dataset = AccentSameSentenceDataset(dataset_path, input_accent, target_accent, sr, n_fft, hop_length, max_frames, n_mels)
data_loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

model = AccentMorpherCNN().to(device)
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-3)
loss_fn = nn.L1Loss()

for epoch in range(epochs):
    model.train()
    total_loss = 0
    for X, Y in tqdm(data_loader, desc=f"Epoch {epoch+1}"):
        X, Y = X.to(device), Y.to(device)
        optimizer.zero_grad()
        pred = model(X)
        loss = loss_fn(pred, Y)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    avg_loss = total_loss / len(data_loader)
    print(f"Epoch {epoch+1}: Loss = {avg_loss:.4f}")

# --- SAVE MODEL ---
torch.save(model.state_dict(), "accent_morpher_stronger.pth")
```

```python
import os
import torch
import numpy as np
import librosa
from IPython.display import Audio

# --- Constants ---
dataset_path = "/content/drive/MyDrive/recordings"
input_accent = "arabic"
sr = 16000
n_fft = 512
hop_length = 256
n_mels = 80

# --- Load model definition again ---
class AccentMorpherCNN(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = torch.nn.Sequential(
            torch.nn.Conv2d(1, 32, kernel_size=3, padding=1),
            torch.nn.BatchNorm2d(32),
            torch.nn.ReLU(),
            torch.nn.Conv2d(32, 64, kernel_size=3, padding=1),
            torch.nn.BatchNorm2d(64),
            torch.nn.ReLU(),
        )
        self.bottleneck = torch.nn.Sequential(
            torch.nn.Conv2d(64, 64, kernel_size=1),
            torch.nn.ReLU()
        )
        self.decoder = torch.nn.Sequential(
            torch.nn.Conv2d(64, 32, kernel_size=3, padding=1),
            torch.nn.BatchNorm2d(32),
            torch.nn.ReLU(),
            torch.nn.Conv2d(32, 1, kernel_size=3, padding=1)
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.bottleneck(x)
        x = self.decoder(x)
        return x

# --- Load model weights safely ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = AccentMorpherCNN().to(device)
model.load_state_dict(torch.load("accent_morpher_stronger.pth", map_location=torch.device('cpu')))
model.eval()

# --- Load and process input accent audio ---
input_path = os.path.join(dataset_path, 'arabic', 'arabic2.mp3')
wave, _ = librosa.load(input_path, sr=sr)

mel_spec = librosa.feature.melspectrogram(y=wave, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)
log_mel = np.log1p(mel_spec)
test_input = torch.tensor(log_mel, dtype=torch.float32).unsqueeze(0).unsqueeze(0).to(device)

# --- Get model prediction ---
with torch.no_grad():
    pred = model(test_input).squeeze().cpu().numpy()

# --- Postprocess predicted log-mel spectrogram ---
pred_mel = np.expm1(pred)
pred_mel1 = pred_mel / np.max(pred_mel + 1e-8)

# --- Use HiFi-GAN to synthesize audio ---
# Download HiFi-G
hifigan = torch.hub.load(
    'NVIDIA/DeepLearningExamples:torchhub',
    'nvidia_hifigan',
    model_math='fp32'
)
hifigan = hifigan.to(device)
hifigan.eval()

# Prepare mel for HiFi-GAN
mel_tensor = torch.tensor(pred_mel1).unsqueeze(0).to(device)  # [1, n_mels, T]
with torch.no_grad():
    audio_tensor = hifigan(mel_tensor).squeeze().cpu()

audio = audio_tensor.numpy()
Audio(audio, rate=sr)
```

```python
import noisereduce as nr
import librosa
import numpy as np
from pydub import AudioSegment
from pydub.effects import normalize
import soundfile as sf
from IPython.display import Audio


def clean_audio_from_array(y, sr=16000):
    """
    Takes an audio array `y` and sampling rate `sr`,
    returns a cleaned (noise reduced + normalized) audio array.
    """

    # Noise reduction (assume first 0.5 sec is noise)
    noise_sample = y[:int(sr * 0.5)]
    y_denoised = nr.reduce_noise(y=y, sr=sr, y_noise=noise_sample)

    # Save temporarily to normalize (pydub works with files)
    sf.write("temp.wav", y_denoised, sr)

    # Normalize loudness
    audio = AudioSegment.from_wav("temp.wav")
    audio = normalize(audio)

    # Convert back to numpy array
    samples = np.array(audio.get_array_of_samples()).astype(np.float32)
    if audio.channels == 2:
        samples = samples.reshape((-1, 2)).mean(axis=1)  # convert stereo to mono

    samples = samples / np.iinfo(audio.array_type).max  # normalize to -1 to 1 range

    return samples
    Audio(samples, rate=sr)
```
